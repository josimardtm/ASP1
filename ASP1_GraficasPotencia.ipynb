```python
import plotly.graph_objs as go
import plotly.io as pio
from plotly.subplots import make_subplots

# Configurar la visualización en modo interactivo
#pio.renderers.default = "notebook"
pio.renderers.default = "colab"

# Gráfica de voltaje y corriente instantáneos
t=np.linspace(0,4/60,1000)
delta=0
beta=-36.8699*(np.pi/180)
v=220*np.sqrt(2)*np.cos(2*np.pi*60*t+delta)
i=113.636*np.sqrt(2)*np.cos(2*np.pi*60*t+beta)
p=v*i
pr=220*113.636*np.cos(delta-beta)*(1+np.cos(2*(2*np.pi*60*t+delta)))
px=220*113.636*np.sin(delta-beta)*(np.sin(2*(2*np.pi*60*t+delta)))

# Crear las trazas para cada variable
traza_voltaje = go.Scatter(x=t, y=v, mode='lines+markers', name='Voltaje (V)', marker=dict(size=2))
traza_corriente = go.Scatter(x=t, y=i, mode='lines+markers', name='Corriente (A)', marker=dict(size=2))
traza_potencia = go.Scatter(x=t, y=p, mode='lines+markers', name='Potencia (W)', marker=dict(size=2))
traza_potencia_activa = go.Scatter(x=t, y=pr, mode='lines+markers', name='Potencia Real (W)', marker=dict(size=2))
traza_potencia_reactiva = go.Scatter(x=t, y=px, mode='lines+markers', name='Potencia Reactiva (VAr)', marker=dict(size=2))

# Crear la figura
#fig = go.Figure()
fig = make_subplots(rows=2, cols=1, shared_xaxes=True,subplot_titles=("Voltaje y Corriente vs. Tiempo", "Potencia vs. Tiempo"))

# Añadir las trazas a la figura
fig.add_trace(traza_voltaje, row=1, col=1)
fig.add_trace(traza_corriente, row=1, col=1)
fig.add_trace(traza_potencia, row=2, col=1)
fig.add_trace(traza_potencia_activa, row=2, col=1)
fig.add_trace(traza_potencia_reactiva, row=2, col=1)

# Configurar el layout
fig.update_layout(
    title='Voltaje, Corriente y Potencia en función del Tiempo',
    xaxis_title='Tiempo (s)',
    yaxis_title='Valores',
    legend_title='Variables',
    hovermode='x'
)

# Mostrar la figura
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Gráfica de voltaje y corriente instantáneos
t=np.linspace(0,4/60,1000)
delta=0
beta=-31.788*(np.pi/180)
v=220*np.sqrt(2)*np.cos(2*np.pi*60*t+delta)
i=15*np.sqrt(2)*np.cos(2*np.pi*60*t+beta)
p=v*i
results={
    't':t,
    'v':v,
    'i':i,
    'p':p
}

results = pd.DataFrame(results)

fig = px.line(results, x="t", y="v", labels={'x':'t', 'y':'v(t)'})
fig.add_scatter(x=results['t'], y=results['p'], mode='lines',showlegend=False)
fig.show()
```
